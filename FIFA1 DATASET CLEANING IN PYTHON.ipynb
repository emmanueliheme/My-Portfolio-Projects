## Objective: write a function that cleans the data following the below guidelines
### Project Guide lines
1. Do the height and weight columns have the appropriate data types?
2. Can you separate the joined column into year, month and day column?
3. Can you clean and transform the value, wage and release clause columns into columns of integers?
4. How can you remove the newline characters from the hits column?
5. Should you separate the Team & Contract column into separate team and contract columns?


```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\Users\HomePC\Downloads\fifa21_raw_data.csv")
df2 = pd.read_csv(r"C:\Users\HomePC\Downloads\fifa21 raw data v2.csv")
```

```python
pd.set_option('display.max_columns', None)
```

```python
# Merge only unique columns from df2 (excluding those already in df1 except the key)
common_cols = df1.columns.intersection(df2.columns).tolist()
key = 'ID'

cols_to_add = [col for col in df2.columns if col not in common_cols or col == key]

df3 = pd.merge(df1, df2[cols_to_add], on=key)

df3.head()
```

```python
def Clean_dataset(df3):
    #Drop  columns
    df3.drop(columns=["photoUrl", "Club", "playerUrl", "Loan Date End", "Contract"], inplace=True) 

    #remove newline character \n from columns
    df3['Hits'] = df3['Hits'].str.replace('\n', '', regex=False) 
    df3['Team & Contract']=df3['Team & Contract'].str.replace('\n', '', regex=False)

    # Use regex to separate into two columns
    df3[['Team', 'Contract']] = df3['Team & Contract'].str.extract(r'^(.*?)(\d{4}\s*~\s*\d{4})$')
    df3['Contract'] = df3['Contract'].str.replace('~', '-', regex=False) 
    df3.drop(columns=["Team & Contract"], inplace=True)
    df3.head()

    # Clean and convert 
    df3['Value_int'] = (
        df3['Value']
        .str.replace('€', '', regex=False)
        .str.replace('M', '', regex=False)
        .str.replace('K', '', regex=False)
        .astype(float)
        .mul(1_000_000)
        .astype(int)
    )    
    df3['Wage_int'] = (
        df3['Wage']
        .str.replace('€', '', regex=False)
        .str.replace('K', '', regex=False)
        .astype(float)
        .mul(1000)
        .astype(int)
    )
    df3['Release Clause_int'] = (
        df3['Release Clause']
        .str.replace('€', '', regex=False)
        .str.replace('M', '', regex=False)
        .str.replace('K', '', regex=False)
        .astype(float)
        .mul(1_000_000)
        .astype(int)
    )
    #Remove string 'lbs' and convert weight to integer
    df3['Weight_int'] = df3['Weight'].str.replace('lbs', '', regex=False).astype(int)

    # Extract feet and inches using regex
    df3[['Feet', 'Inches']] = df3['Height'].str.extract(r"(\d+)'(\d+)?")

    # Convert to integers and compute total height in inches
    df3['Height_in'] = df3['Feet'].astype(int) * 12 + df3['Inches'].astype(int)

    #Convert inches to cm and drop Feet, inches columns
    #df3['Height_cm'] = df3['Height_in'] * 2.54

    #drop Feet and Inches column
    df3.drop(columns=["Feet", "Inches"], inplace=True)

    #Split Joined column into Year, Month and Day columns
    df3[["Joined", "Year"]] = df3['Joined'].str.split(',', expand=True)
    df3[["Month", "Day"]] = df3['Joined'].str.split(' ', expand=True)
    
    return df3
```

```python
df3_cleaned = Clean_dataset(df3)
```
